# Une question : Voici une liste d'employés en full remote. La société XCoinCoin souhaite définir un lieu de rendez-vous en présentiel.
#Quel serait le meilleur point de rencontre ? Pourquoi ?
#Représenter aussi l'ensemble sur une carte

```python
import pandas as pd
import requests
import folium

df = pd.DataFrame({"employé": ["emma", "noah", "sophia", "liam", "olivia", "william", "ava", "james", "isabella", "benjamin"],
                           "ville": ["paris", "marseille", "lyon", "toulouse", "nice", "nantes", "strasbourg", "montpellier", "bordeaux", "lille"]})
```

```python
def point(ville):

    link_main = 'https://api-adresse.data.gouv.fr/search/?q='

    link = link_main + ville

    r = requests.get(link).json()
    loc = r['features'][0]['geometry']['coordinates'][::-1]

    return loc
```

```python
m = folium.Map(location=point('Vernais'),zoom_start=5.5)
```

```python
df['point'] = df['ville'].apply(point)
```

```python
for n in range(len(df)):
  folium.Marker(
      location=df['point'].iloc[n],
      popup= df['employé'].iloc[n]
      ).add_to(m)
```

```python
df['longitude'] = df['point'].apply(lambda x : x[0])
df['latitude'] = df['point'].apply(lambda x : x[1])
```

```python
point_moyen = [df['longitude'].mean(), df['latitude'].mean()]
point_median = [df['longitude'].median(), df['latitude'].median()]

folium.Marker(
    location=point_moyen,
    popup= 'Point_moyen',
    icon=folium.Icon(color='red')
    ).add_to(m)

folium.Marker(
    location=point_median,
    popup= 'Point_median',
    icon=folium.Icon(color='red')
    ).add_to(m)

m
```

```python
'''
Le meilleur point de rencontre serait aux alentours de Clermont-Ferrand qui est la plus grande ville
proche des points moyens et médians de chaque employé.
Sinon à Lyon, car une employée y habite et c'est la ville habitée par un employé la plus proche des points médian et moyen.
'''
```
